#### b. Understanding how many new customers acquired every month

```python
# Combining orders and customer table

cust_orders = Orders.merge( Customers, how = "inner", on = "customer_id" )
cust_orders.head()
```

```python
cust_orders.order_purchase_timestamp = cust_orders.order_purchase_timestamp.apply(lambda x: pd.to_datetime(x,format="%m/%d/%Y %H:%M"))
cust_orders.order_approved_at = cust_orders.order_approved_at.apply(lambda x: pd.to_datetime(x,format="%m/%d/%Y %H:%M"))
cust_orders.order_delivered_carrier_date = cust_orders.order_delivered_carrier_date.apply(lambda x: pd.to_datetime(x,format="%m/%d/%Y %H:%M"))
cust_orders.order_delivered_customer_date = cust_orders.order_delivered_customer_date.apply(lambda x: pd.to_datetime(x,format="%m/%d/%Y %H:%M"))
cust_orders.order_estimated_delivery_date = cust_orders.order_estimated_delivery_date.apply(lambda x: pd.to_datetime(x,format="%m/%d/%Y %H:%M"))
```

```python
cust_orders["Year"] = cust_orders.order_purchase_timestamp.apply(lambda x: x.year)
cust_orders["Month"] = cust_orders.order_purchase_timestamp.apply(lambda x: x.month)
cust_group = cust_orders.groupby(["Year","Month","customer_id"])["order_id"].count().reset_index().sort_values(["Year","Month"])
cust_mnth_yr = cust_group.groupby(["Year","Month"])["customer_id"].count().reset_index()
```

```python
cust_group.head()
```

```python
cust_mnth_yr.head()
```

```python
#  b. Understanding how many new customers acquired every month

cust_old = set()

for i,j in cust_mnth_yr.iterrows():
    df = (cust_group [( cust_group.Year == j.Year ) & (cust_group.Month == j.Month) ])
    
    total_count = len(set(df.customer_id))
    new_count = len(set(df.customer_id).difference(cust_old))
    cust_old = cust_old.union(set(df.customer_id))
    cust_mnth_yr.loc[i,"total_count"] = int(total_count)
    cust_mnth_yr.loc[i,"new_count"] = int(new_count)
    
cust_mnth_yr
```

#### c. Understand the retention of customers on month on month basis

```python
cust_mnth_yr["cust_retained"] = cust_mnth_yr.total_count - cust_mnth_yr.new_count

cust_mnth_yr
```

### 2. Performing Customers/sellers Segmentation


#### a. Divide the customers into groups based on the revenue generated


```python
 cust_seg = business_data[["customer_id", "payment_value"]]

cust_seg = cust_seg.groupby(["customer_id"])["payment_value"].sum().reset_index()
cust_seg
```

```python
cust_seg.quantile(q = [0.25, 0.5,0.75])
```

```python
cust_seg.payment_value.min()
```

```python
cust_seg.payment_value.max()
```

```python
def segment(pay):
        if pay <= 62.84:
            return "Low Spender"
        elif pay <= 110.13:
            return "Medium Spender"
        elif pay <= 196.58:
            return "High Spender"
        else:
            return "Extravagant"    
```

```python
cust_seg["cust_category"] = cust_seg.payment_value.apply(segment)
```

```python
cust_seg.cust_category
```
